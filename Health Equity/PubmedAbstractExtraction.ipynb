```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load the CSV file
csv_file = "/content/drive/MyDrive/Equity Project/data.xlsx"  # Replace with the path to your CSV file
data = pd.read_excel(csv_file)


def get_pubmed_abstract(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, "html.parser")
        # Find the abstract in the HTML
        abstract = soup.find("div", {"class": "abstract-content"}).text.strip()
        return abstract
    except Exception as e:
        return f"Error: {str(e)}"

# Add a new column for the abstract
data['Abstract'] = data['Pubmed link'].apply(lambda x: get_pubmed_abstract(x) if pd.notna(x) else "No link provided")

# Save the updated data to a new CSV
output_file = "updated_file_with_abstracts.csv"
data.to_csv(output_file, index=False, sep='\t')

print(f"Abstracts added and saved to {output_file}")
```
